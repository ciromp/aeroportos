```python
from bs4 import BeautifulSoup
from prettytable import PrettyTable
import requests
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 8)
print("\n\nA aplicação lista os vôos de três aeroportos, montando uma tabela comparativa ao final da quantidade de vôos partindo de cada aeropoto.")
print("\n\n O sistem funciona de forma universal, bastando apenas trocar as abreviações nas três variáveis na célula abaixo.")
```

```python
#apreviação dos aeroportos utilizados
abreviacao1 = "GRU"
abreviacao2 = "BSB"
abreviacao3 = "GIG"

url = "http://www.flightstats.com/go/weblet?guid=120d3d56318c8a24:-10d0d2d1:145dc93a67b:-16bc&weblet=status&action=AirportFlightStatus&airportCode=" + abreviacao1

response = requests.get(url, headers={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"});
content = response.content

soup = BeautifulSoup(content, "html.parser")

data_partida = []
table_partida = soup.find('table', attrs={'class':'tableListingTable'})

rows = table_partida.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data_partida.append([ele for ele in cols if ele]) # Get rid of empty values
    
    
tabela_voos_partida = PrettyTable(["Vôo", "Companhia", "Destino", "Horário", "Status"])
for line in data_partida[1:]:
    tabela_voos_partida.add_row(line)
```

```python
print("\n\n>> Partidas do Aeroporto: " + abreviacao1 + "\n\n")
print(tabela_voos_partida)
```

```python
import plotly.plotly as py

import pandas as pd
import numpy as np

total = [[abreviacao1,float(len(data_partida))],[abreviacao2,float(len(data_partida_2))],[abreviacao3,float(len(data_partida_3))]]
print("\n\n >> Quantidade de vôos partindo dos aeroportos analisados abaixo \n\n")
df = pd.DataFrame(total, columns=["Aeroporto", "Total"])
df.head()
```

```python
print("\n\n >> Gráfico de comparação da quantidade de vôos dos aeroportos analisadaos \n\n")
partidas_aeroporto = df.groupby(['Aeroporto'])["Total"].sum()

grafico = partidas_aeroporto.plot(kind='bar',legend=None,title="Quantidade de vôos partindo por aeroporto.")
grafico.set_xlabel("Aeroportos")
grafico.set_ylabel("Quantidade de vôos partindo")
```
